```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = './loss/val_avg'
csv_file_list = os.listdir(path)
```

```python
for csv_file in csv_file_list:
    file_path = path + '/{}'.format(csv_file)
    name = path + '/{}.csv'.format(csv_file[:-4].split('_')[-1])
    os.rename(file_path, name)
    
```

```python
version = ['loss', 'loss_determinants', 'loss_points']

for i in range(len(version)):    
    train_path = './loss/train_{}_avg'.format(version[i])
    train_loss_list = os.listdir(train_path)

    val_path = './loss/validation_{}_avg'.format(version[i])
    val_loss_list = os.listdir(val_path)

    train_losses = []
    val_losses = []
    iter = []
    # for num, loss_file in enumerate(train_loss_list):
    train_loss = np.loadtxt(train_path + '/{}'.format('23.csv'), delimiter=',')
    val_loss = np.loadtxt(val_path + '/{}'.format('23.csv'), delimiter=',')
    # train_losses.append(train_loss)
    # val_losses.append(val_loss)
    iter = list(range(0, len(train_loss)))

    plt.plot(iter, train_loss)
    plt.plot(iter, val_loss)

    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    # plt.title('Experiment Result')
    plt.legend(['train_{}_avg'.format(version[i]), 'validation_{}_avg'.format(version[i])])
    plt.show()
```
